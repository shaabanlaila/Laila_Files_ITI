# Merging Allocs Files from ITI Files
Taking multipe allocs files from iti.files Qgroundcontrol Allocs folder and merging them together

Needed to run some extra code to assure columns match up correctly with their values and to remove extra column. Also had to remove filler 14 from alloc_ip column.

```python
# imports here
import pandas as pd
import numpy as np
```

```python
mergeddf = pd.concat( 
    map(pd.read_csv, ['v20.6.6.3.enable2.QGroundControl.allocs',
    'v20.6.6.3.enable1.QgroundControl.allocs',
    'v20.6.1.60.enable2.QgroundControl.allocs',
    'v20.6.1.60.enable1.QgroundControl.allocs',
    'v20.6.1.38.enable2.Qgroundcontrol.allocs',
    'v20.6.1.38.enable1.Qgroundcontrol.allocs',
    'Sept14.r13.run2.QgroundControl.allocs',
    'Sept14.r13.run1.QgroundControl.allocs',
    'Sept13.r12.QgroundControl.allocs',
    'Oct14.r11.3.Qgroundcontrol.allocs',
    'Nov7.r11.3.QgroundControl.allocs',
    'Nov6.r17.QgroundControl.allocs',
    'notfromFreshBoot.v20.6.6.3.enable2.QGroundControl.allocs',
    '6.6.3.v20.23.10.distro.enable1.QgroundControl.allocs'])).reset_index()
mergeddf.head()
```

```python
df = mergeddf
df.head()
```

```python
df = df.rename(columns = {'index': 'alloc_ip','alloc_ip': 'pid','pid': 'ptr','ptr': 'size','size': 'alloc_serial','alloc_serial': 'free_serial','free_serial': 'free_ip','free_ip': 'type','type': 'command','slab':'slab','new_type':'new_type'})
df.head()
```

```python
df = df.drop(df[df['alloc_ip'] == '14              '].index)
df
```

```python
print('Number of Rows:', df.shape[0])
print('Number of Columns:', df.shape[1])
```

```python
df = df.copy()
df['alloc_jiffies'].fillna(df['alloc_serial'], inplace=True)
df['free_ip'].fillna(df['free_jiffies'], inplace=True)
df = df.drop(columns = ['alloc_serial','free_jiffies'])
df = df.rename(columns = {'free_serial':'free_jiffies'})
df.head()
```

```python
df.to_csv('onlinefilesdatabase.csv')
```
