# Merging Allocs Files from Hard Drive
Taking multipe allocs files from allocs hard drive folder and merging them together

Needed to run some extra code to assure columns match up correctly with their values and to remove extra column. Also had to remove filler 14 from alloc_ip column.

```python
# imports here
import pandas as pd
import numpy as np
```

```python
# merging all files from allocs folder
mergeddf = pd.concat( 
    map(pd.read_csv, ['v20.6.6.3.enable2.KillAllLess.allocs',
    'v20.6.1.60.enable2.KillAllLess.allocs',
    'v20.6.1.38.enable2.less-killall.allocs',
    '6.6.3.v20.23.10.distro.enable2.KillAllLess.allocs',
    'v20.6.1.60.enable1.KillAllLess.allocs',
    'v20.6.1.38.enable1.less-killall.allocs',
    'v20.6.6.3.enable2.QGroundControl.allocs',
    '6.6.3.v20.23.10.distro.enable2.QgroundControl.allocs',
    'v20.6.1.60.enable2.QgroundControl.allocs',
    '6.6.3.v20.23.10.distro.enable1.QgroundControl.allocs',
    'notfromFreshBoot.v20.6.6.3.enable2.QGroundControl.allocs',
    'v20.6.6.3.enable1.QgroundControl.allocs',
    'v20.6.1.38.enable2.Qgroundcontrol.allocs',
    'v20.6.1.60.enable1.QgroundControl.allocs',
    'v20.6.1.38.enable1.Qgroundcontrol.allocs'])).reset_index()
mergeddf
```

```python
df = mergeddf
df.head()
```

```python
df = df.rename(columns = {'index': 'alloc_ip','alloc_ip': 'pid','pid': 'ptr','ptr': 'size','size': 'alloc_jiffies','alloc_serial': 'free_jiffies','free_serial': 'free_ip','free_ip': 'type','type': 'command','slab':'slab','new_type':'new_type'})
df.head()
```

```python
df = df.drop(df[df['alloc_ip'] == '14              '].index)
df.head()
```

```python
print('Number of Rows:', df.shape[0])
print('Number of Columns:', df.shape[1])
```

```python
df.to_csv('folderdatabase.csv')
```
